```python
from sklearn.inspection import plot_partial_dependence
from bank_mlp_model import load_model
from bank_dataset_preprocessing import load_preprocessed_data

mlp_classifier = load_model()
X, y, X_test, X_train, y_test, y_train = load_preprocessed_data()

print('Computing partial dependence plots...')
features = ['pdays', 'cons.price.idx']
display = plot_partial_dependence(mlp_classifier, X_train, features, kind="both", subsample=50,
       n_jobs=3, grid_resolution=50, random_state=0)
display.figure_.suptitle(
    'pdays (number of days since last contact) and CPI \n'
    'for the bank-telemarketing dataset, with MLPClassifier'
)
display.figure_.subplots_adjust(hspace=0.3)
```
