```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import shap
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import confusion_matrix, classification_report, plot_roc_curve
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn import set_config
from sklearn.impute import SimpleImputer

data = pd.read_csv("data/bank-additional/bank-additional.csv", sep=";")
```

```python
# numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
# categorical_features = data.select_dtypes(include='object').drop(['y'], axis=1).columns

numerical_features = ['age', 'campaign', 'duration', 'pdays', 'previous', 'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
categorical_features_nominal = ['job', 'marital', 'contact', 'month', 'day_of_week']
categorical_features_binary = ['default', 'housing', 'loan']
categorical_features_unknown_values = ['education', 'default', 'housing', 'loan', 'job', 'marital']

X = data.drop(['y', 'poutcome'], axis=1)
y = data['y']

# smt = SMOTE(random_state=0)

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.33, random_state=0)

education_ordinal_mapping = {'illiterate': 0, 'basic.4y': 1, 'basic.6y': 2, 'basic.9y': 3, 'high.school': 4, 'professional.course': 5,
                             'university.degree': 6}
binary_features_mapping = {'yes': 1, 'no': 0}

# handle missing unknown values
simple_imputer = SimpleImputer(missing_values="unknown", strategy="most_frequent")
simple_imputer.fit(X_train[categorical_features_unknown_values])
X_train[categorical_features_unknown_values] = simple_imputer.transform(X_train[categorical_features_unknown_values])
X_test[categorical_features_unknown_values] = simple_imputer.transform(X_test[categorical_features_unknown_values])

# Categorical to ordinal
X_train['education'] = X_train['education'].replace(education_ordinal_mapping)
X_test['education'] = X_test['education'].replace(education_ordinal_mapping)

# Encoding binary values
X_train[categorical_features_binary] = X_train[categorical_features_binary].replace(binary_features_mapping)
X_test[categorical_features_binary] = X_test[categorical_features_binary].replace(binary_features_mapping)

# Encoding predicted value
y_train = y_train.replace(binary_features_mapping)
y_test = y_test.replace(binary_features_mapping)

# Scaling
scalar = StandardScaler()
scalar.fit(X_train[numerical_features])

X_train[numerical_features] = scalar.transform(X_train[numerical_features])
X_test[numerical_features] = scalar.transform(X_test[numerical_features])

# One hot encoding categorical nominal variables
X_train_dummies = pd.get_dummies(X_train[categorical_features_nominal])
X_test_dummies = pd.get_dummies(X_test[categorical_features_nominal])
# Dropping
X_train = X_train.drop(categorical_features_nominal, axis='columns')
X_test = X_test.drop(categorical_features_nominal, axis='columns')
# Concatenating
X_train = pd.concat([X_train, X_train_dummies,], axis='columns')
X_test = pd.concat([X_test, X_test_dummies,], axis='columns')

print("Wasup my dudes")
```
